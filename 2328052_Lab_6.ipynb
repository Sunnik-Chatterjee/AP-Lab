Q1. Matrix multiplication and time comparison

```python
import time
import numpy as np

A = [[i for i in range(100)] for j in range(100)]
B = [[i for i in range(100)] for j in range(100)]

start = time.time()
C = [[0]*100 for _ in range(100)]
for i in range(100):
    for j in range(100):
        for k in range(100):
            C[i][j] += A[i][k]*B[k][j]
end = time.time()
print("List Time:", end-start)

A1 = np.array(A)
B1 = np.array(B)

start = time.time()
C1 = np.dot(A1,B1)
end = time.time()
print("NumPy Time:", end-start)
```

Q3. Class for 2D array operations

```python
class Array2D:
    def __init__(self,a):
        self.a=a
    def row_reverse(self):
        return [row[::-1] for row in self.a]
    def col_reverse(self):
        return self.a[::-1]
    def both_reverse(self):
        return [row[::-1] for row in self.a[::-1]]
    def cut_border(self):
        return [row[1:-1] for row in self.a[1:-1]]

a = Array2D([[1,2,3],[4,5,6],[7,8,9]])
print(a.row_reverse())
print(a.col_reverse())
print(a.both_reverse())
print(a.cut_border())
```

Q4. Solve linear equations using inverse

```python
import numpy as np

# (i)
A1 = np.array([[1,2],[2,4]])
B1 = np.array([-4,-8])

if np.linalg.det(A1)==0:
    print("System (i): Singular Matrix → Infinite Solutions")
else:
    X1 = np.linalg.solve(A1,B1)
    print("System (i) Solution:", X1)

# (ii)
A2 = np.array([[1,2,3],[2,-3,2],[3,1,-1]])
B2 = np.array([6,14,-2])

X2 = np.linalg.solve(A2,B2)
print("System (ii) Solution:", X2)
```

Q6) Write a function to flatten a 2D array into 1D array without using flatten() or ravel().

```python
def flat(a):
    b=[]
    for i in a:
        for j in i:
            b.append(j)
    return b

print(flat([[1,2],[3,4]]))
```
