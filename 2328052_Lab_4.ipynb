Q1. Create sets and perform union, intersection, difference

```python
s1=set(map(int,input().split()))
s2=set(map(int,input().split()))
print("Set1:",s1)
print("Set2:",s2)
print("Union:",s1|s2)
print("Intersection:",s1&s2)
print("Difference:",s1-s2)
```

Q2. Check subset

```python
a=set(map(int,input().split()))
b=set(map(int,input().split()))
print(a.issubset(b))
```

Q3. Square of common elements of two sets

```python
a=set(map(int,input().split()))
b=set(map(int,input().split()))
c=set()
for i in a&b:
    c.add(i*i)
print(c)
```

Q4. Create immutable set and add new element

```python
fs=frozenset(map(int,input().split()))
print(fs)
fs2=fs.union({int(input())})
print(fs2)
```

Q6. Sort dictionary and remove a key

```python
d={'b':2,'a':1}
d=dict(sorted(d.items()))
print(d)
d.pop(input())
print(d)
```

Q7. Word frequency

```python
w=input().split()
d={}
for i in w:
    d[i]=d.get(i,0)+1
print(d)
```

Q8. Factorial using function

```python
def fact(n):
    f=1
    for i in range(1,n+1):
        f*=i
    return f
print(fact(int(input())))
```

Q9. Sum of list using recursion

```python
def s(l):
    if not l:
        return 0
    return l[0]+s(l[1:])
print(s(list(map(int,input().split()))))
```

Q10. Square and cube using function

```python
def sc(n):
    return n*n,n*n*n
for i in map(int,input().split()):
    print(sc(i))
```

Q11. Sort list using lambda

```python
l=list(map(int,input().split()))
print(sorted(l,key=lambda x:x))
```
