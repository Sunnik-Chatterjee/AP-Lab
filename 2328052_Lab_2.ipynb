Q1. Write Python program to read an alphabet from the user and convert it into
opposite case (upper to lower or lower to upper case).

```python
word=input("Enter a word with any case i would opposite it: ")
for i in word:
  if(i==i.lower()): print(i.upper(),end='')
  else: print(i.lower(),end='')
```

Q2. Write Python program to generate Fibonacci series up to specified no. of terms.

```python
n=int(input("Upto which term you want your fibonacci series: "))
print(0,end=' ')
print(1,end=' ')
term=3
i=1
j=0
while(term<=n):
  num = i+j
  print(num,end=' ')
  j=i
  i=num
  term+=1
```

Q3. Write Python program to find out the prime factors of a number entered through
keyboard (distinct).

```python
num = int(input("Enter a number: "))
if(n<=1): print("No prime factors")
ans=""
i=2
while(num%i==0):
  ans+=str(i)
  num//=i
i=3
while(i*i<=num):
  while(num%i==0):
    ans+=str(i)
    num//=i
  i+=2
if(num>1): ans+=str(num)
i=1
print(ans[0])
while(i<len(ans)):
  if(ans[i]==ans[i-1]): i+=1
  else:
    print(ans[i])
    i+=1
```

Q4. Write Python program to input a decimal number, then convert it into equivalent
number as follows using switch case statement:
1) Binary number
2) Octal number
3) Hexadecimal number

```python
num = int(input("Enter a number: "))
print("1. Binary number")
print("2. Octal number")
print("3. Hexadecimal number")
choice=int(input("Enter the number of choice of what formate you want the number to be shown: "))
match choice:
  case 1:
    binary_output = bin(num)
    print(binary_output)
  case 2:
    octal_output=oct(num)
    print(octal_output)
  case 3:
    hexaDecimal_output = hex(num)
    print(hexaDecimal_output)
```

Q5. Create a guessing game where the user has to guess a randomly generated
number. Use branching, looping, and flow control statements to manage the game&#39;s
flow.

```python
import random

target = random.randint(1, 100)
attempts = 0

while True:
    guess = int(input("Guess a number (1-100): "))
    attempts += 1
    if guess == target:
        print(f"Correct! Took {attempts} guesses.")
        break
    print("Too low." if guess < target else "Too high.")
```
