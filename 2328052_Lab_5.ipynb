1. Create class Box with length, width, height and print volume.

```python
class Box:
    def __init__(self,l,w,h):
        self.l=l
        self.w=w
        self.h=h
    def volume(self):
        return self.l*self.w*self.h

l=int(input())
w=int(input())
h=int(input())
b=Box(l,w,h)
print("Output: ",b.volume())
```

2. Create Employee class and inherit Manager and Developer.

```python
class Employee:
    companyName="TechCorp"
    def __init__(self,name,age,exp,dept,salary):
        self.name=name
        self.__age=age
        self.exp=exp
        self.dept=dept
        self.__salary=salary
    def show(self):
        print(self.name,self.exp,self.dept)

class Manager(Employee):
    def __init__(self,name,age,exp,dept,salary,teams):
        super().__init__(name,age,exp,dept,salary)
        self.teams=teams
    def totalTeams(self):
        return len(self.teams)

class Developer(Employee):
    def __init__(self,name,age,exp,dept,salary,lang):
        super().__init__(name,age,exp,dept,salary)
        self.lang=lang
    def showLang(self):
        print(self.lang)

m=Manager("Amit",35,10,"IT",80000,["A","B"])
d=Developer("Riya",25,3,"IT",40000,"Python")

m.show()
print(m.totalTeams())
d.show()
d.showLang()
```

3. Create Binary Tree with add node, count nodes, height.



```python
class Node:
    def __init__(self,val):
        self.val=val
        self.left=None
        self.right=None

class BinaryTree:
    def __init__(self):
        self.root=None

    def add(self,val):
        if not self.root:
            self.root=Node(val)
            return
        q=[self.root]
        while q:
            temp=q.pop(0)
            if not temp.left:
                temp.left=Node(val)
                return
            else:
                q.append(temp.left)
            if not temp.right:
                temp.right=Node(val)
                return
            else:
                q.append(temp.right)

    def count(self,node):
        if not node:
            return 0
        return 1+self.count(node.left)+self.count(node.right)

    def height(self,node):
        if not node:
            return 0
        return 1+max(self.height(node.left),self.height(node.right))

t=BinaryTree()
t.add(1)
t.add(2)
t.add(3)
print(t.count(t.root))
print(t.height(t.root))
```

4. Create Holiday class with name, day, month.
Check same month, find average day, and create Independence Day object.

```python
class Holiday:
    def __init__(self,name,day,month):
        self.name=name
        self.day=day
        self.month=month

    def inSameMonth(self,h):
        return self.month==h.month

    @staticmethod
    def avgDate(arr):
        s=0
        for h in arr:
            s+=h.day
        return s/len(arr)

h1=Holiday("Independence Day",15,"August")
h2=Holiday("Republic Day",26,"January")
h3=Holiday("Gandhi Jayanti",2,"October")

print(h1.inSameMonth(h2))
print(Holiday.avgDate([h1,h2,h3]))
print(h1.name,h1.day,h1.month)
```

Create Movie class with title, studio, rating.
Default rating = PG. Return only PG movies. Create Casino Royale object.

```python
class Movie:
    def __init__(self,title,studio,rating="PG"):
        self.title=title
        self.studio=studio
        self.rating=rating

    @staticmethod
    def getPG(arr):
        res=[]
        for m in arr:
            if m.rating=="PG":
                res.append(m.title)
        return res

m1=Movie("Casino Royale","Eon Productions","PG-13")
m2=Movie("Toy Story","Pixar","PG")
m3=Movie("Avengers","Marvel","PG-13")

print(Movie.getPG([m1,m2,m3]))
print(m1.title,m1.studio,m1.rating)
```
