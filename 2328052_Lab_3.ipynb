largest and smallest number from a list. Also find sum of all the items in a list.

Q4. Write a Python program that takes a list and prints a new list with unique elements of
the first list.

```python
myList=list()
uniqueList=list()
i=int(input("Enter how many numbers you want to enter: "))
for i in range(0,i):
  a=int(input("Enter the number: "))
  myList.append(a)
for i in myList:
  if(i not in uniqueList):
    uniqueList.append(i)
print(uniqueList)
```

Q.10) Given a list of integers, write a Python program to find the longest increasing
subsequence. The function should return the length of the subsequence and the subsequence
itself. Ensure that the subsequence is a continuous subsequence (not necessarily unique).

```python
def lis(a):
    if not a: return 0, []
    m = cur = 1
    s = 0
    for i in range(1, len(a)):
        if a[i] > a[i-1]:
            cur += 1
        else:
            if cur > m:
                m = cur
                s = i - cur
            cur = 1
    if cur > m:
        m = cur
        s = len(a) - cur
    return m, a[s:s+m]

arr = [10, 12, 3, 4, 5, 20, 21, 22, 1, 2]
print(lis(arr))
```
